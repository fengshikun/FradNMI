```python
import altair as alt
import copy
import pandas as pd


log_dir = '/home/fengshikun/Pretraining-Denoising/log/time_step_2'
file_name = 'data_0.csv'
file_dir = f'{log_dir}/{file_name}'

def read_energy_csv(log_dir, idx=0):
    energy_step = None
    file_names = [f'data_{idx}.csv', f'data_rdkit_opt_True_{idx}.csv', f'data_rdkit_opt_False_{idx}.csv', f'data_{idx}_add_noise.csv']
    for i, f in enumerate(file_names):
        energy_lst = pd.read_csv(f'{log_dir}/{f}')
        energy_lst.rename(columns = {'#"Step"':'Step'}, inplace = True)
        energy_lst.rename(columns = {'Potential Energy (kJ/mole)': f'energy_{i}'}, inplace = True)
        if energy_step is None:
            energy_step = energy_lst[['Step', f'energy_{i}']]
        else:
            energy_step[f'energy_{i}'] = energy_lst[f'energy_{i}']
    return energy_step

energy_step = read_energy_csv(log_dir)
# energy_step
# alt.Chart(energy_step).mark_line(point=True,interpolate="natural").encode(
#     alt.X('Step:Q'
#     ),
#     alt.Y('Potential Energy (kJ/mole):Q'
#     )
# ).interactive()

# df_long = pd.melt(energy_step, id_vars='Step', var_name='y_values')


# alt.Chart(df_long).mark_line(point=True,interpolate="natural").encode(
#     x=alt.X('Step', title='Steps'),
#     y=alt.Y('y_values', title='Energy'),
#     color=alt.Color('y_values', title='Curve', scale=alt.Scale(scheme='category10')),
#     strokeDash=alt.StrokeDash('y_values', title='energy', legend=None, 
#                               scale=alt.Scale(domain=['energy_0', 'energy_1', 'energy_2']))
# ).properties(title='Energy per Steps').interactive()

alt.Chart(energy_step).transform_fold(['energy_0', 'energy_1', 'energy_2', 'energy_3']).mark_line().encode(
    x='Step:Q',
    y='value:Q',
    color='key:N'
).interactive()
```
